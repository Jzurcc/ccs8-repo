```python
import pandas as pd

df_raw = pd.read_csv('happiness.csv')
df_raw.head()
```

```python
from scipy.stats import shapiro

for col in df_raw.columns:
  if col != 'Country name':
    stat, p = shapiro(df_raw[col])
    # print(f'{col}: {p}')
    if p > 0.05:
      print("Likely Normal Distribution (Fail to reject H0)")
    else:
      print("Not Normal Distribution (Reject H0)")
```

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

# Load dataset from Orange

# Convert Orange Table to NumPy array
X = np.array(df.drop('Positive affect', axis=1))  # Features
y = np.array(df['Positive affect']).ravel()  # Target variable

# Define the Random Forest model
rf = RandomForestRegressor(random_state=42)

# Define hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Perform GridSearchCV
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='neg_root_mean_squared_error', n_jobs=-1)
grid_search.fit(X, y)

# Get the best model and its parameters
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
best_rmse = np.sqrt(-grid_search.best_score_)

# Print results
print(f"Best Parameters: {best_params}")
print(f"Best RMSE: {best_rmse}")

# Output the best model
out_model = best_model
```
